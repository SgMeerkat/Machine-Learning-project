```python
import pandas as pd
```

```python
# create for loop to gather and scrape required data from multiple web pages
years = range(1997,2022)

Data = {}
for y in years:
  year = y
  # URL to scrape, notice f string:
  Data["url" + str(y)] = f"https://hoopshype.com/salaries/players/{year}-{year+1}/"
```

```python
Data
```

```python
season_from = 1997 # for season column
season_to = 98 # for season column
x = 0
y = 0

pl_Data = {}

for i in Data:
  if season_to == 97 or season_to == 98: # for years 1997-98 and 1998-99
    pl_Data["pl" + str(season_from + 1)] = pd.read_html(Data[i])[0]
 

  elif season_to not in range(10,23): # for years 1999-00 through 2008-09
    season_to = 0 + x
    pl_Data["pl" + str(season_from + 1)] = pd.read_html(Data[i])[0]
    x += 1

  else: # for years 2009-10 through 2021-22
    season_to = 10 + y
    pl_Data["pl" + str(season_from + 1)] = pd.read_html(Data[i])[0]
    y += 1
  season_from += 1    
  season_to += 1
```

```python
pl_Data
```

```python
# creating the data frame with all tables gathered from webpages
NBA_Player_DF = pd.concat(pl_Data)
NBA_Player_DF = NBA_Player_DF.drop_duplicates()
NBA_Player_DF.reset_index(drop=True, inplace=True)
```
